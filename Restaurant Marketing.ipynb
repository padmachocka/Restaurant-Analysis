# Explain the factors in the data that may have an effect on ratings e.g. No. of cuisines, cost, delivery option etc.

```python
#Analyzing correlations helps identify key factors that influence restaurant ratings, 
#such as number of cuisines, cost, votes and Price range.
#This analysis can guide improvements in service and offerings to boost ratings.
```

```python
correlation_matrix = merged_table1[['Aggregate rating', 'No. of Cuisines', 'Average Cost for two', 'Votes', 'Price range']].corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix of Factors Affecting Ratings')
plt.show()
```

```python
#Cuisine diversity and Price range influence restaurant ratings.
```
